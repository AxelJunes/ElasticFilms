```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
# Get data frame with data from movies
df = pd.read_csv('movies.csv', sep=',')
```

```python
df.head(5)
```

```python
# Get number movies dependign on genre
df["genres"].value_counts()
```

```python
# Save only western movies to new dataframe
western_movies = df.loc[df['genres'] == 'Western']
```

```python
western_movies
```

```python
# Get movie IDs
western_titles = western_movies.as_matrix(columns=df.columns[1:2])
```

```python
from imdb import IMDb
ia = IMDb()
movieIDs = []
for title in western_titles:
    movies = ia.search_movie(title.item(0))
    if(len(movies) > 0):
        movieIDs.append(movies[0].movieID) # Get ids from the first film that matches the query
```

```python
# Create a clean data frame with movie titles and plots
western_df = pd.DataFrame(index=np.arange(0, len(movieIDs)), columns=['title', 'plot'])
```

```python
from imdb import IMDb
ia = IMDb()
i = 0
for m_id in movieIDs:
    m = ia.get_movie(m_id)
    western_df.loc[i]['title'] = m.get('title')
    western_df.loc[i]['plot'] = m.get('plot outline')
    i += 1
```

```python
western_df = western_df.dropna(axis=0)
```
